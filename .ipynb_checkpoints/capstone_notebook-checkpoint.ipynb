<h3>Now that all the libraries we will need have been imported, let's scrape our data from the Wikipedia page and store it in a Pandas dataframe.</h3>

```python
url = "https://en.wikipedia.org/wiki/List_of_neighborhoods_in_Chicago"
page = request.urlopen(url)

#feed the HTML data into a Beautiful Soup object
soup = BeautifulSoup(page, 'html.parser')

#uncomment the following line to view the raw HTML data
#print(soup.prettify())
```

```python
#extract the table from the HTML file
html_data = soup.find('table', class_ = 'wikitable sortable')

#uncomment the follow line to view the raw HTML table data
#html_data
```

In HTML, rows are signified by the \<tr> tag and columns are signified by the \<th> tag. The data itself is signified with the \<td> tag. Let's loop through the table one row at a time with the findAll() function provided by Beautiful Soup. From observing the table on the Wikipedia page, we know the table has 2 columns: Neighborhood and Community Area.

```python
A = []
B = []

for row in html_data.findAll('tr'):
    cells = row.findAll('td')
    if len(cells) == 2:
        #add the data to the appropriate list and strip away the \n for each data point
        A.append(cells[0].find(text = True).rstrip('\n'))
        B.append(cells[1].find(text = True).rstrip('\n'))
```

```python
#store the data in a Pandas dataframe
df = pd.DataFrame(A, columns = ["Neighborhood"])
df["Community Area"] = B

#view the dataframe
df.head(10)
```

```python
df.shape
```

Reading the shape of the dataframe, df, we see there are 246 neighborhoods in Chicago, per the Wikipedia page.

<h3>Now that we have successfully stored the list of Chicago neighborhoods in a Pandas dataframe, let's find the coordinates of each neighborhood. We will need these coordinates for the Foursquare API and plotting later on.</h3>

```python
#define a function for retrieving the latitude and longitude coordinates of a neighborhood
def get_coords(neighborhood):
    coords = None
    
    while (coords is None):
        g = geocoder.arcgis('{}, Chicago, United States'.format(neighborhood))
        coords = g.latlng
    #return the coordinates
    return coords
```

```python
#create new list to hold the coordinates of each neighborhood
neigh_lat_long = [ get_coords(neighborhood) for neighborhood in df["Neighborhood"].tolist() ]
```

```python
print(len(neigh_lat_long))    #to confirm the number of coordinate pairs matches neighborhoods
```

```python
#populate the coordinate pairs in a Pandas dataframe
df_coords = pd.DataFrame(neigh_lat_long, columns = ['Latitude', 'Longitude'])
df_coords.head()
```

```python
#merge the coordinates information with the original Neighborhood dataframe
df['Latitude'] = df_coords['Latitude']
df['Longitude'] = df_coords['Longitude']

#also drop the Community Area column, since it is not needed
df.drop(['Community Area'], axis = 1, inplace = True)
print(df.shape)
```

<p>Now, let's get nearby venues for each neighborhood from Foursquare. <span style = "font-weight:bold;">Be careful how many times you run this function. The Foursquare API only allows 950 API calls per day for a basic account. Make too many calls and the program will return a 'groups' error.</span></p>

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&query={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION,
            search_query,
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
chicago_venues = getNearbyVenues(names = df['Neighborhood'], latitudes = df['Latitude'],
                                longitudes = df['Longitude'])
```

```python
#use one-hot encoding to set up k-means clustering
venues_onehot = pd.get_dummies(chicago_venues[['Venue Category']], prefix = "", prefix_sep = "")

# add neighborhood column back to dataframe
venues_onehot['Neighborhoods'] = chicago_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [venues_onehot.columns[-1]] + list(venues_onehot.columns[:-1])
venues_onehot = venues_onehot[fixed_columns]

print(venues_onehot.shape)
venues_onehot.head()
```

<p>Let's group the data by neighborhood and find out each neighborhood's most popular venues.</p>

```python
chicago_grouped = venues_onehot.groupby('Neighborhoods').mean().reset_index()
chicago_grouped.head(10)
```

```python
print(chicago_grouped.shape)
```

<h3>Cluster</h3>

```python
#choose to have 3 clusters
k = 3
chicago_cluster = chicago_grouped.drop('Neighborhoods', axis = 1)

kmeans = KMeans(n_clusters = k, random_state = 0).fit(chicago_cluster)
kmeans.labels_[0:10]
```

```python
labels = kmeans.labels_
```

```python
chicago_final = chicago_grouped.copy()
chicago_final['Cluster'] = labels
chicago_final.drop(chicago_final.columns.difference(['Neighborhoods','Cluster']), 1, inplace=True)
chicago_final.head()
```

```python
chicago_final.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
chicago_final.head()
```

<p>Each neighborhood has been successfully assigned a cluster from the k-means algorithm. Let's merge the coordinates obtained earlier with the cluster data for each neighborhood. This will allow us to plot the clusters on a map.</p>

```python
chicago_final = chicago_final.merge(df, on="Neighborhood")

print(chicago_final.shape)
chicago_final.head()
```

```python
#Let's group the neighborhoods by cluster 
chicago_final.sort_values(['Cluster'], inplace = True)
chicago_final.head()
```

<h3>Finally, plot the clusters on a map using Folium</h3>

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i+x+(i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(chicago_final['Latitude'], chicago_final['Longitude'], chicago_final['Neighborhood'], chicago_final['Cluster']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
